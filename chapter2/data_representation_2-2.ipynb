# Section 2.2: Data presentation (scalar, vector, matrix and tensor)

```python
import numpy as np
```

```python
#scalar - rank 0 tensor
x = np.array(12)
print(x)
print(x.ndim)
```

```python
x = np.array([12, 3, 6, 14, 7])
print(x)
print(x.ndim)
```

```python
x = np.array([
    [5, 78, 2, 34, 0],
    [6, 79, 3, 35, 1],
    [7, 80, 4, 36, 2]
])
print(x.ndim)
```

```python
x = np.array([[[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
              [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
              [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]]])
print(x.ndim)
```

Three key attributes of a tensor:  
    - Dimension  
    - Shape  
    - Data type

```python
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print("train_image dimensions: {}, shape: {}, dtype: {}".format(train_images.ndim, train_images.shape, train_images.dtype))
```

#### Display the 4th digit in train_images:

```python
digit = train_images[4]
import matplotlib.pyplot as plt
plt.imshow(digit, cmap = plt.cm.binary)
plt.show()
print("label: {}".format(train_labels[4]))
```

#### Try tensor slicing

```python
myslice = train_images[10:100]
print(myslice.shape)
```

```python
myslice = train_images[10:100, :, :]
print(myslice.shape)
myslice = train_images[10:100, 0:28, 0:28]
print(myslice.shape)
```

#### Notion of data batches:  
The way we create the nth batch of 128 images  
When considering a batch tensor, first axis (axis 0) is called batch axis or batch dimension. This term will be frequently used in Keras or other DL libraries

```python
n = 3
batch = train_images[n*128 : (n + 1)*128]
print(batch.shape)
```
